```python
import pandas as pd
```

```python
df1 = pd.DataFrame({
    "city": ["new york","chicago","orlando"],
    "temperature": [21,14,35],
})
df1
```

```python
df2 = pd.DataFrame({
    "city": ["chicago","new york","orlando"],
    "humidity": [65,68,75],
})
df2
```

```python
merged = pd.merge(df1,df2,on='city')
merged
```

```python
df3 = pd.DataFrame({
    "city": ["new york","chicago","orlando", "baltimore"],
    "temperature": [21,14,35, 38],
})
df3
```

```python
df4 = pd.DataFrame({
    "city": ["chicago","new york","san diego"],
    "humidity": [65,68,71],
})
df4
```

```python
merged1 = pd.merge(df3,df4,on='city',how='inner')
merged1
```

```python
merged2 = pd.merge(df3,df4,on='city',how='outer')
merged2
```

```python
merged3 = pd.merge(df3,df4,on='city',how='left')
merged3
```

```python
merged4= pd.merge(df3,df4,on='city',how='right')
merged4
```

```python
df5 = pd.DataFrame({
    "city": ["new york","chicago","orlando", "baltimore"],
    "temperature": [21,14,35,38],
    "humidity": [65,68,71, 75]
})
df5
```

```python
df6 = pd.DataFrame({
    "city": ["chicago","new york","san diego"],
    "temperature": [21,14,35],
    "humidity": [65,68,71]
})
df6
```

```python
merged5 = pd.merge(df5,df6,on='city',how='outer',suffixes=('left','right'))
merged5
```
